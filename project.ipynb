## 12. For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.

```python
list(loans)
```

considero solo il caso in cui planned_expiration_time è maggiore o uguale a disburse_time e elimino i valori nulli

```python
loans['planned_expiration_time']=pd.to_datetime(loans['planned_expiration_time'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
loans['disburse_time']=pd.to_datetime(loans['disburse_time'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
#considero solo i casi in cui expiration è maggiore o uguale a disburse
loans=loans.loc[(loans['planned_expiration_time']>=loans['disburse_time'])]
#elimino i valori nulli
loans=loans[np.isfinite((loans['planned_expiration_time'])&(loans['disburse_time']))]
loans[['planned_expiration_time','disburse_time','loan_amount']].head()
```

```python
diz={}
for (expiration, disburse, amount) in zip(loans['planned_expiration_time'],loans['disburse_time'], loans['loan_amount']):
    year_expiration=int(expiration.year)
    year_disburse=int(disburse.year)
    difference=(expiration-disburse).days
    if year_expiration == year_disburse:
        if year_expiration in diz:
            diz[year_expiration]+=amount
        else:
            diz[year_expiration]=amount
    else:
        #.item serve per estrarre solo il numero
        #freq='D' per calendar day frequency
        days_in_first_year = 365-disburse.dayofyear
        ris=amount*days_in_first_year/difference
        if year_disburse in diz:
            diz[year_disburse]+=ris
        else:
            diz[year_disburse]=ris
        days_in_last_year =expiration.dayofyear
        ris2=amount*days_in_last_year/difference
        if year_expiration in diz:
            diz[year_expiration]+=ris2
        else:
            diz[year_expiration]=ris2
        if (year_expiration-year_disburse)>1:
            ris3=365*amount/difference
            for year in range(year_disburse,year_expiration):
                if year in diz:
                    diz[year]+=ris3
                else:
                    diz[year]=ris3
print(diz)
```
